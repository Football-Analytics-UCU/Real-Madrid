```python
import pandas as pd
import plotly.graph_objects as go
```

```python
timing_chart_df = pd.read_csv('./data/xg_timing_chart.csv')
```

```python
# Create traces for each of the teams xG
trace0 = go.Scatter(
    x = timing_chart_df[timing_chart_df['h_a'] == 'Real Madrid']['minute'],
    y = timing_chart_df[timing_chart_df['h_a'] == 'Real Madrid']['running_sum_xG'],
    mode = 'lines',
    name = 'Real Madrid',
    line = dict(
        color = ('#E5E64B'),
        width = 4)
)
trace1 = go.Scatter(
    x = timing_chart_df[timing_chart_df['h_a'] == 'Barcelona']['minute'],
    y = timing_chart_df[timing_chart_df['h_a'] == 'Barcelona']['running_sum_xG'],
    mode = 'lines',
    name = 'Barcelona',
    line = dict(
        color = ('#00BCD4'),
        width = 4)
)

data_comp = [trace0, trace1]

layout_comp = go.Layout(
    autosize=False,
    width=800,
    height=600,
    title='Timing Chart',
    plot_bgcolor='#3E3E40',
    hovermode='x',
    xaxis=dict(
        title='Minute',
        ticklen=15,
        zeroline=True,
        showgrid=True,
        gridcolor='#39393B',
        gridwidth=2,
    ),
    yaxis=dict(
        title='xG',
        ticklen=5,
        gridwidth=2,
        zeroline=True,
        showgrid=True,
        gridcolor='#39393B',
    ),
)

fig_comp = go.Figure(data=data_comp, layout=layout_comp)
fig_comp.show()
```
