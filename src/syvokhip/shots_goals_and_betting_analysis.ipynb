```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
```

```python
df = pd.read_csv('LaLigaOdds23-24.csv')

rma_matches = df[((df['HomeTeam'] == 'Real Madrid')|(df['AwayTeam'] == 'Real Madrid'))][['Date', 'Time', 'HomeTeam', 'AwayTeam', 'FTR', 'FTHG', 'FTAG', 'HS', 'AS', 'HST', 'AST', 'MaxH', 'MaxD', 'MaxA', "AvgH", 'AvgD', 'AvgA', 'Max>2.5', 'Max<2.5', 'Avg>2.5', 'Avg<2.5']]
gir_matches = df[((df['HomeTeam'] == 'Girona')|(df['AwayTeam'] == 'Girona'))][['HomeTeam', 'AwayTeam', 'FTR', 'FTHG', 'FTAG', "AvgH", 'AvgD', 'AvgA',]]
bar_matches = df[((df['HomeTeam'] == 'Barcelona')|(df['AwayTeam'] == 'Barcelona'))][['HomeTeam', 'AwayTeam', 'FTR', 'FTHG', 'FTAG', "AvgH", 'AvgD', 'AvgA']]

rma_matches
```

# Real Madrid Data Processing

Setting up our dataframe, calculating new values and adding new metrics for matches

```python
rma_matches['profit'], rma_matches['totalOverProfit'], rma_matches['totalUnderProfit'] = -100, -100, -100
rma_matches['shotsOnTarget'], rma_matches['shots'], rma_matches['rmaGoals'], rma_matches['shotsPercentage'], rma_matches['totalGoals'], rma_matches['total_win'], rma_matches['matchweek']= 0, 0, 0, 0, 0, 0, 0
rma_matches['isWin'], rma_matches['isHome'] = True, True

matchdays = [int((i+1)) for i in range(len(rma_matches.index))]

rma_matches['shotsPercentage'] = np.where((rma_matches['HomeTeam'] == 'Real Madrid'), rma_matches['HST']/rma_matches['HS'], rma_matches['shotsPercentage'])
rma_matches['shotsPercentage'] = np.where((rma_matches['AwayTeam'] == 'Real Madrid'), rma_matches['AST']/rma_matches['AS'], rma_matches['shotsPercentage'])
rma_matches['totalGoals'] = rma_matches['FTHG'] + rma_matches['FTAG']
rma_matches['shots'] = np.where((rma_matches['HomeTeam'] == 'Real Madrid'), rma_matches['HS'] , rma_matches['shots'])
rma_matches['shots'] = np.where((rma_matches['AwayTeam'] == 'Real Madrid'), rma_matches['AS'] , rma_matches['shots'])
rma_matches['rmaGoals'] = np.where((rma_matches['HomeTeam'] == 'Real Madrid'), rma_matches['FTHG'] , rma_matches['rmaGoals'])
rma_matches['rmaGoals'] = np.where((rma_matches['AwayTeam'] == 'Real Madrid'), rma_matches['FTAG'] , rma_matches['rmaGoals'])
rma_matches['shotsOnTarget'] = np.where((rma_matches['HomeTeam'] == 'Real Madrid'), rma_matches['HST'] , rma_matches['shotsOnTarget'])
rma_matches['shotsOnTarget'] = np.where((rma_matches['AwayTeam'] == 'Real Madrid'), rma_matches['AST'] , rma_matches['shotsOnTarget'])
rma_matches['isWin'] = np.where((rma_matches['HomeTeam'] != 'Real Madrid') & (rma_matches['FTR'] == 'H'), False, rma_matches['isWin'])
rma_matches['isHome'] = np.where((rma_matches['HomeTeam'] != 'Real Madrid'), False, rma_matches['isHome'])

rma_matches['matchweek'] = np.where((rma_matches['profit'] == rma_matches['profit'].values), matchdays , rma_matches['total_win'])
rma_matches['matchweek'] = rma_matches['matchweek'].astype(int)

rma_matches
```

# Processing and calculating of shots and goals data.

Visualization of shots accuracy(the ratio of shots on target to the total amount of shots) and
goal conversion rate(the ratio of goals scored to the total amount of total shots on target)

```python
#conversion_rate = sum(rma_matches['rmaGoals'].values) / sum(rma_matches['shots'].values)
#goals_per_game = sum(rma_matches['rmaGoals'].values) / len(rma_matches['matchweek'])
#shots_accuracy = sum(rma_matches['shotsOnTarget'].values) / sum(rma_matches['shots'].values)
sns.set(rc={'figure.figsize':(10,5)})


shots = rma_matches[['shots', 'shotsOnTarget', 'rmaGoals', 'shotsPercentage']].reset_index(drop = True)
shots['conversionRate'] = rma_matches['rmaGoals'].values / rma_matches['shots'].values
shots['matchweek'] = matchdays

ax = sns.lineplot(data = shots, x = 'matchweek', y = 'shotsPercentage', marker = '*', label = 'Shots Accuracy')
sns.lineplot(data = shots, x = 'matchweek', y = 'conversionRate', marker = '*', label = 'Shots Conversion').set(
    xlabel = 'Matchweek', ylabel = 'Rate', title = 'Shots accuracy compared to goal conversion rate')

ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
ax.xaxis.set_major_formatter(ticker.ScalarFormatter())

ax.fill_between(x = shots['matchweek'],
                y1 = shots['shotsPercentage'], y2 = 0, alpha = 0.3, facecolor = 'green')

ax.fill_between(x = shots['matchweek'],
                y1 = shots['conversionRate'], y2 = 0, alpha = 0.3, facecolor = 'orange')

ax.axhline(shots['conversionRate'].mean(), color = 'black', label = 'mean', linestyle="dashed")
ax.text(x = shots['matchweek'].mean(), y = shots['conversionRate'].mean(),
        s = 'mean: {:.3f}'.format(shots['conversionRate'].mean())).set_backgroundcolor('white')

ax.axhline(shots['shotsPercentage'].mean(), color = 'black', label = 'mean', linestyle="dashed")
ax.text(x = shots['matchweek'].mean(), y = shots['shotsPercentage'].mean(),
        s = 'mean: {:.3f}'.format(shots['shotsPercentage'].mean())).set_backgroundcolor('white')
#shots = pd.melt(shots, value_vars=['shots', 'shotsOnTarget', 'rmaGoals'],
#             var_name='outcome_type', value_name='amount')
```

# Betting odds processing for TOP3 teams in LaLiga season 23/24

General betting odds calculations for events such as team win/loss and 
number of goals scored in the match (over 2.5 and under 2.5)

**Each calculation was made assuming that the amount of the bet was 100$

```python
rma_matches['profit'] = np.where(((rma_matches['HomeTeam'] == 'Real Madrid') & (rma_matches['FTR'] == 'H')), (100 * rma_matches['AvgH'] - 100), rma_matches['profit'])
rma_matches['profit'] = np.where(((rma_matches['AwayTeam'] == 'Real Madrid') & (rma_matches['FTR'] == 'A')), (100 * rma_matches['AvgA'] - 100), rma_matches['profit'])

matches_profit = np.cumsum(rma_matches['profit'].values)

rma_matches['total_win'] = np.where((rma_matches['profit'] == rma_matches['profit'].values), matches_profit , rma_matches['total_win'])

rma_matches['totalOverProfit'] = np.where((rma_matches['totalGoals'] >= 3), (100 * rma_matches['Avg>2.5'] - 100), rma_matches['totalOverProfit'])
rma_matches['totalUnderProfit'] = np.where((rma_matches['totalGoals'] < 3), (100 * rma_matches['Avg<2.5'] - 100), rma_matches['totalUnderProfit'])

home_matches = rma_matches[(rma_matches['HomeTeam'] == 'Real Madrid')][['HomeTeam', 'AwayTeam', 'FTR', 'FTHG', 'FTAG', 'profit', 'isWin', 'total_win' , 'shotsPercentage', 'shots', 'shotsOnTarget', 'rmaGoals', 'totalGoals', 'matchweek', 'totalOverProfit', 'totalUnderProfit']]
away_matches = rma_matches[(rma_matches['AwayTeam'] == 'Real Madrid')][['HomeTeam', 'AwayTeam', 'FTR', 'FTHG', 'FTAG', 'profit', 'isWin', 'total_win' ,'shotsPercentage', 'shots', 'shotsOnTarget', 'rmaGoals', 'totalGoals', 'matchweek', 'totalOverProfit', 'totalUnderProfit']]

matches_before_mw11 = rma_matches[rma_matches['matchweek'] < 11]
```
